```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
# Load the dataset
loan_data = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/basic-dataset/refs/heads/master/Loan-Approval-Prediction.csv')

# Split features and target
```

```python
loan_data.head()
```

Create a copy to convert to numerical value

```python
loan_data_numerical= loan_data.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Convert categorical columns to numerical
loan_data_numerical['Gender'] = le.fit_transform(loan_data_numerical['Gender'])
loan_data_numerical['Married'] = le.fit_transform(loan_data_numerical['Married'])
loan_data_numerical['Education'] = le.fit_transform(loan_data_numerical['Education'])
loan_data_numerical['Self_Employed'] = le.fit_transform(loan_data_numerical['Self_Employed'])
loan_data_numerical['Property_Area'] = le.fit_transform(loan_data_numerical['Property_Area'])
loan_data_numerical['Loan_Status'] = le.fit_transform(loan_data_numerical['Loan_Status'])
loan_data_numerical['Dependents'] = le.fit_transform(loan_data_numerical['Dependents'])
```

Remove na values

```python
loan_data_numerical =loan_data_numerical.dropna()
```

```python
loan_data_numerical.info()
```

Convert to X an y, split data to training and testing

```python

X= loan_data_numerical.drop('Loan_Status', axis=1)
X.drop('Loan_ID', axis=1, inplace=True)  # Drop Loan_ID as it's not a feature
y= loan_data_numerical['Loan_Status']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,       # 20% for testing, 80% for training (common ratio)
    random_state=42,     # Ensures the split is the same every time you run the code
    stratify=y           # Ensures the class distribution in the splits matches the original data
)
```

Train and evaluate Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score, auc

# Create an instance of the LogisticRegression model
model = LogisticRegression(solver='liblinear') # 'liblinear' is a good default solver

# Fit the model to the training data
model.fit(X_train, y_train)
```

Evaluate Linear model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report

# ... (Assume model is already trained and X_test, y_test are available from step 1)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate various classification metrics
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Precision (weighted): {precision_score(y_test, y_pred, average='weighted')}")
print(f"Recall (weighted): {recall_score(y_test, y_pred, average='weighted')}")
print(f"F1 Score (weighted): {f1_score(y_test, y_pred, average='weighted')}")

# Display a full classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

fpr, tpr, thresholds = roc_curve(y_test, y_pred)

auc_score = roc_auc_score(y_test, y_pred)
print(f'ROC AUC Score: {auc_score:.4f}')


# Display a confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

Train and Evaluate KNN model

```python
from sklearn.neighbors import KNeighborsClassifier

# Create an instance of the KNeighborsClassifier model
knn_model = KNeighborsClassifier(n_neighbors=5)  # Default k=5, can be tuned

# Fit the model to the training data
knn_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_knn = knn_model.predict(X_test)

# Calculate various classification metrics
print(f"Accuracy: {accuracy_score(y_test, y_pred_knn)}")
print(f"Precision (weighted): {precision_score(y_test, y_pred_knn, average='weighted')}")
print(f"Recall (weighted): {recall_score(y_test, y_pred_knn, average='weighted')}")
print(f"F1 Score (weighted): {f1_score(y_test, y_pred_knn, average='weighted')}")

# Display a full classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_knn))

# Calculate ROC AUC score
auc_score_knn = roc_auc_score(y_test, y_pred_knn)
print(f'ROC AUC Score: {auc_score_knn:.4f}')

# Display a confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_knn))
```
